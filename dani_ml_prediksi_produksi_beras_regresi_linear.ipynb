###a. Menentukan masalah bisnis
**Judul Penelitian:**
Prediksi Produksi Beras Berdasarkan Produksi Padi Menggunakan Regresi Linear

**Masalah Bisnis:**
Produksi beras merupakan faktor krusial bagi ketahanan pangan Indonesia. Namun, fluktuasi dalam produksi padi sering kali menyulitkan perencanaan produksi beras secara nasional dan daerah. * Distribusi beras tidak merata.
* Lonjakan harga karena kekurangan pasokan.
* Kegagalan dalam mencapai target swasembada beras nasional.

**Tujuan Penelitian:**
Membangun model prediksi produksi beras berdasarkan data produksi padi menggunakan algoritma regresi linear. Model ini akan membantu:
* Pemerintah atau dinas pertanian memprediksi produksi beras lebih awal.
* Mengoptimalkan distribusi dan logistik.
* Membantu perumusan kebijakan pertanian secara lebih berbasis data.

**Kriteria Kesuksesan:**
* Model regresi linear dibangun dengan akurasi tinggi.
* Model mampu memprediksi produksi beras dari input produksi padi untuk provinsi-provinsi di Indonesia dengan galat prediksi rendah (RMSE kecil).
* Model mudah digunakan kembali oleh stakeholder (misalnya: dinas pertanian atau pengambil kebijakan).
* Hasil visualisasi dan prediksi dapat membantu proses pengambilan keputusan.

###b. Menentukan tugas analitik untuk menyelesaikan masalah

**Jenis Tugas Analitik:**
Regresi

**Penjelasan:**

* Target (label/output) yaitu Produksi Beras bersifat numerik dan kontinu.
* Fitur (input) yaitu Produksi Padi tersedia dalam dataset.

##**C. Praprocessing Data**

Pada bagian praprocessing data, akan dilakukan pembersihan data untuk bisa diolah pada tahap selanjutnya. Pembersihan yang dimaksud adalah menangani data-data seperti data duplikat, data null, dan data outlier.

####**Tujuan dan Alasan Pemilihan Fitur:**

Dalam studi kasus "Prediksi Produksi Beras Berdasarkan Produksi Padi", maka:

* Fitur utama (X): Produksi Padi - Produksi Padi (ton) (Ton) — sebagai input/variabel prediktor.

* Target (y): Produksi Padi - Produksi Beras (ton) (Ton) — sebagai output/variabel yang ingin diprediksi.

Tidak memakai kolom "Provinsi" karena:

* Data numeriknya tidak dipengaruhi secara langsung oleh nama provinsi.

* Model regresi linear fokus pada hubungan numerik antar fitur, bukan kategori daerah.

```python
# Rename kolom agar lebih mudah digunakan
df.rename(columns={
    'Produksi Padi - Produksi Padi (ton) (Ton)': 'Produksi_Padi',
    'Produksi Padi - Produksi Beras (ton) (Ton)': 'Produksi_Beras'
}, inplace=True)

df.head(10)
```

**Cek dan drop data duplikat**

Pada kode di bawah, terlihat bahwa data duplikat yang ada adalah data yang disisipkan oleh pembuat dataset berupa baris kosong dan teks catatan di bawah akhir dataset.

```python
# Menampung data duplikat
duplicates_boolean_series = df.duplicated(keep=False)
duplicated_rows = df[duplicates_boolean_series]

# Cetak baris-baris yang duplikat
print("Baris data yang terduplikasi:")
print(duplicated_rows)

print("\nJumlah baris sebelum menghapus data duplikat:", len(df))
print("\nJumlah baris yang merupakan data duplikat:", duplicates_boolean_series.sum())

# Hapus data duplikat
df.drop_duplicates(inplace=True)

# Cek jumlah
print("\nJumlah baris setelah menghapus data duplikat:", len(df))
```

**Cek dan drop data null (NAN)**

Seperti pada penjelasan sebelumnya, dataset memiliki beberapa baris data yang NAN, hal ini karena terjadinya penyisipan baris kosong dan catatan di baris paling bawah, juga ada beberapa provinsi yang tidak memiliki data produksi padi atau beras. Bisa dilihat pada penampilan di bawah.

```python
df.tail(10)
```

```python
print("Jumlah data null per kolom:")
df.isnull().sum()
```

Penanganan data null (NAN)

```python
# Menghapus data null berdasarkan kolom
df_clean = df.dropna(subset=['Produksi_Padi', 'Produksi_Beras'])

print("\nJumlah baris sebelum menghapus data null:", len(df))
print("\nJumlah baris setelah menghapus data null:", len(df_clean))

# Cek ukuran data sebelum dan sesudah
print("\nUkuran sebelum cleaning:", df.shape)
print("Ukuran sesudah cleaning null:", df_clean.shape)
```

**Penanganan Data Outlier**

Ditangani dengan mendeteksi nilai outlier dan kemudian melakukan filter data.

```python
# Cek outlier dengan IQR (Interquartile Range)
Q1_padi = df_clean['Produksi_Padi'].quantile(0.25)
Q3_padi = df_clean['Produksi_Padi'].quantile(0.75)
IQR_padi = Q3_padi - Q1_padi

Q1_beras = df_clean['Produksi_Beras'].quantile(0.25)
Q3_beras = df_clean['Produksi_Beras'].quantile(0.75)
IQR_beras = Q3_beras - Q1_beras

# Batas atas & bawah
batas_bawah_padi = Q1_padi - 1.5 * IQR_padi
batas_atas_padi = Q3_padi + 1.5 * IQR_padi

batas_bawah_beras = Q1_beras - 1.5 * IQR_beras
batas_atas_beras = Q3_beras + 1.5 * IQR_beras

# Filter data tanpa outlier
df_final = df_clean[
    (df_clean['Produksi_Padi'] >= batas_bawah_padi) &
    (df_clean['Produksi_Padi'] <= batas_atas_padi) &
    (df_clean['Produksi_Beras'] >= batas_bawah_beras) &
    (df_clean['Produksi_Beras'] <= batas_atas_beras)
]

print("\nUkuran data final (tanpa outlier):", df_final.shape)
```

```python
# Dataset sudah bersih dan siap pakai untuk tahap selanjutnya
df_final
```

**Visualisasi data final**

Grafik korelasi produksi padi dan produksi beras yang sudah clean atau memakai data final.

```python
# Grafik Korelasi Produksi Padi dan Produksi Beras
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df_final,
    x='Produksi_Padi',
    y='Produksi_Beras'
)
plt.title("Scatter Plot: Produksi Padi vs Produksi Beras")
plt.xlabel("Produksi Padi (ton)")
plt.ylabel("Produksi Beras (ton)")
plt.tight_layout()
plt.show()
```

Menampilkan distribusi data final dengan histogram

```python
# Grafik Distribusi Produksi Padi
plt.figure(figsize=(8,5))
plt.subplot(1, 2, 1)
sns.histplot(df_final['Produksi_Padi'], bins=30, kde=True)
plt.title("Distribusi Produksi Padi (ton)")
plt.xlabel("Produksi Padi (ton)")
plt.ylabel("Frekuensi")
plt.tight_layout()

# Grafik Distribusi Produksi Beras
plt.subplot(1, 2, 2)
sns.histplot(df_final['Produksi_Beras'], bins=30, kde=True)
plt.title("Distribusi Produksi Beras (ton)")
plt.xlabel("Produksi Beras (ton)")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```

**Pengecekan terakhir**

### Mencoba menerapkan model

```python
from sklearn.linear_model import LinearRegression

# Split data x dan y
X = df_final[['Produksi_Padi']]
y = df_final['Produksi_Beras']

# Penerapan model LinearRegression()
model = LinearRegression()
model.fit(X, y)

# Prediksi model
df_final['Prediksi_Beras'] = model.predict(X)

# Plot
plt.figure(figsize=(10,6))
plt.scatter(df_final['Produksi_Padi'], df_final['Produksi_Beras'], color='blue', label='Data Aktual')
plt.plot(df_final['Produksi_Padi'], df_final['Prediksi_Beras'], color='red', label='Regresi Linear')
plt.xlabel('Produksi Padi (ton)')
plt.ylabel('Produksi Beras (ton)')
plt.title('Regresi Linear: Produksi Padi vs Produksi Beras')
plt.legend()
plt.grid(True)
plt.show()

# Hasil model
print("Intercept:", model.intercept_)
print("Koefisien:", model.coef_[0])
```
